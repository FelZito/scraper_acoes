```python
import requests
import pandas as pd
```

```python
url = 'https://finance.yahoo.com/most-active'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
site = requests.get(url, headers=headers)
```

```python
dados = pd.read_html(site.text)
df = pd.DataFrame(dados[0])
df
```

```python
df.info()
```

```python
# Removendo coluna inútil

df.drop('52 Week Range', axis=1, inplace=True)
```

```python
# Salvando em excel
excel = pd.ExcelWriter('acoes_yahoo.xlsx', engine='xlsxwriter')
df.to_excel(excel, sheet_name='ACOES')
excel.save()
```
